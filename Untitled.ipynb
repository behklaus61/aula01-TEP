```python
alts = []
altsF = []

for i in range(2):
   
    while True: 
        try:
            altura = float (input(f"{i+1} - Digite uma altura: "))
            break
        except ValueError:
            print("Digite uma altura válida")
            
    while True:
        sexo = input(f"{i+1} - Digite o sexo (M/F): "). upper()
        if sexo in ['M', 'F']:
            break
        else:
            print("Digite um sexo válido (M/F)")

    alts.append(altura)

    if (sexo == 'F'):   
        altsF.append(altura)
        
print (f"A menor altura é: {min(alts)}")
print (f"A maior altura é: {max(alts)}")
print (f"A quantidade de mulheres é: {len(altsF)}")
print (f"A média de altura das mulheres é: {sum(altsF)/len(altsF)}")
```
